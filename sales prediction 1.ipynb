```python
#assigning variables to axes
x=dataset['TV']
y=dataset['Sales']
```

```python
#splitting it into training and testing data
#30% is kept as test data set and remaining 70% is kept as train data set 
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,train_size = 0.7,random_state=100)
```

```python
x_train
```

```python
y_train
```

```python
#to make it 2D data
x_train=np.array(x_train).reshape(-1,1)
x_train
```

```python
#doing the same with test data
x_test=np.array(x_test).reshape(-1,1)
x_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linear_reg=LinearRegression()
```

```python
linear_reg.fit(x_train,y_train)
```

```python
linear_reg.predict(x_test[0].reshape(-1, 1)) 
```

```python
x_test[0]
```

```python
#finding the intercept
c = linear_reg.intercept_
c
```

```python
#finding the coefficient
m = linear_reg.coef_
m
```

The formula for linear regression is y=mx+c
Now we have coefficient(m)  and intercept(c) .So we can estimate the value of y by using the above formula 

```python
#for training dataset
y_pred_traindata = m*x_train+c
y_pred_traindata.flatten()
#flatten() is used to convert the data in 1D
```

```python
#or we can do it directly using predict() formula from sklearn library
y_pred_traindata_1 = linear_reg.predict(x_train)
y_pred_traindata_1
```

Measuring the accuracy of the data by using R square and root mean square error for these we need to import

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
#R square for train data
r2_traindata=(r2_score(y_train,y_pred_traindata_1))
r2_traindata
```

```python
#calculating Root mean squared error (measures the average error the model makes in predicting the outcome)
np.sqrt(mean_squared_error(y_train,y_pred_traindata_1))
```

```python
#visualize the training data using scatter plot
fig = plt.figure(figsize=(6,3))
plt.scatter(x_train,y_train)
plt.plot(x_train,y_pred_traindata_1,color="green")
plt.xlabel("TV")
plt.ylabel("Sales")
plt.show()
```

```python
#Distribution of the error terms to check if the error terms are also normally distributed 
err = (y_train - y_pred_traindata_1)
```

```python
#using histogram to plot the error term
fig = plt.figure(figsize=(6,3))
sns.distplot(err, bins = 15)
fig.suptitle('Error Terms', fontsize = 15)                   
plt.xlabel('y_train - y_pred_traindata_1', fontsize = 15)         
plt.show()
```

```python
#Looking for patterns in the residuals using scatter plot
plt.scatter(x_train,err)
plt.figure(figsize=(6,3))
plt.show()
```

```python
#predict for testing data set
y_pred_test = linear_reg.predict(x_test)
y_pred_test
```

```python
#R square for test data
r2_testdata=r2_score(y_test,y_pred_test)
r2_testdata
```

```python
#calculating Root mean squared error (measures the average error the model makes in predicting the outcome)
np.sqrt(mean_squared_error(y_test,y_pred_test))
```

```python
#visualize the testing data using scatter plot
plt.scatter(x_train,y_train)
plt.plot(x_test,y_pred_test,color="green")
plt.xlabel("TV")
plt.ylabel("Sales")
plt.show()
```
